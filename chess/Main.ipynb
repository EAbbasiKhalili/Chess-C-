```c++
#include <algoviz/SVG.hpp>
#include <iostream>
#include "chess.hpp"
using namespace std;

Rect pixel[8][8];
Circle circles[8][8];
MouseState player;

```

```c++
 //visualizing the chess board
void makingChessBoard(){
    for(int i = 0; i < 8; i++){
        for(int j = 0; j < 8; j++){
            pixel[i][j].moveTo(i*200,j*200);
            pixel[i][j].setWidth(200);
            pixel[i][j].setHeight(200);
            pixel[i][j].addTo(&cBoard);
            pixel[i][j].setColor("black");
            if(j%2 == 0 && i%2 != 0){
                pixel[i][j].setFill(80,27,0);
                
            }
            if(j%2 != 0 && i%2 == 0){
                pixel[i][j].setFill(80,27,0);
            }
            if(j%2 == 0 && i%2 == 0){
                pixel[i][j].setFill(152,118,84);
            }
            if(j%2 != 0 && i%2 != 0){
                pixel[i][j].setFill(152,118,84);
            }
            
        }
    }
}
makingChessBoard();
```

```c++
//using circles for showing each piece avaiable moves
void showYellowCircles(){
    
    for(int i = 0; i < 8; i++){
        for(int j = 0; j < 8; j++){
            circles[i][j].addTo(&cBoard);
            circles[i][j].moveTo(100 + j * 200, 1500 - i * 200);
            circles[i][j].setRadius(50);
            circles[i][j].setFill("yellow");
            }
        }
    
}
showYellowCircles();
```

```c++
//hiding avaiable moves
void hideYellowCircles(){
    
    for(int i = 0; i < 8; i++){
        for(int j = 0; j < 8; j++){
            circles[i][j].hide();
            }
        }
    
}
hideYellowCircles();
```

```c++
//the actual legal moves of each piece is vector of points which we show using circles
void showLegalMove(vector<Point> legalMoves){
    
    for(auto p : legalMoves){
        circles[p.x][p.y].show();
    }

}
```

```c++
//string taste;
int isClicked = 0; 
int pointY = 0;
int pointX = 0;  
int newPointY = 0;
int newPointX = 0;

// actual gameplay
while ( taste != "q" ) {
    //checkmate works after the checked king do one last move although it is not allowed irl.
    if(chess.isBlackCheckMated()==true || chess.isWhiteCheckMated()==true){

        taste = "q";
        message +=  " - checkmate you lost baaaad";
        AlgoViz::sendText(message);
        break;
    }
    
    taste = cBoard.lastKey();
    player = cBoard.lastClick();
    
    if(player.isLegal()){
            
            if(isClicked == 0){
                //selecting the piece
                isClicked = 1;
                pointY = player.x()/200;
                pointX = (1600-player.y())/200;
                
                if(chess.chessBoard[pointX][pointY] != NULL){
                    vector<Point> legalMoves = chess.chessBoard[pointX][pointY]->getLegalMove();
                    showLegalMove(legalMoves);
                }else{
                    isClicked = 0;
                }
                
            }else{
                //selecting the destination
                isClicked = 0;
                newPointY = player.x()/200;
                newPointX = (1600-player.y())/200;
                chess.move(Point{pointX,pointY},Point{newPointX,newPointY});
                hideYellowCircles();
            }
    }
}
```
